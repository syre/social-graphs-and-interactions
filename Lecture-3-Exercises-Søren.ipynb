### Write a program to generate a table of lexical diversity scores (i.e. token/type ratios), as we saw in 1.1. Include the full set of Brown Corpus genres (nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest number of tokens per type)? Is this what you would have expected?

```python
def lexical_diversity(text):
    return len(text) / len(set(text))
# my bet is on religion
for category in brown.categories():
    print("{} category has a lexical diversity score: {:10.2f}".format(category,lexical_diversity(brown.words(categories=category))))
```

###Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
from nltk.corpus import stopwords, inaugural
from nltk import FreqDist
def count_most_occurring(text, number=50):
    """assuming english"""
    stripped = [w.lower() for w in text if w.lower() not in stopwords.words("english") and w.lower().isalnum()]
    f = FreqDist(stripped)
    pprint(f.most_common(number))
count_most_occurring(inaugural.words())
```

###Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
import collections
def count_bigrams(text, number=50):
    """assuming english"""
    bigrams = nltk.bigrams([w.lower() for w in text if w.lower() not in stopwords.words("english") and w.lower().isalnum()])
    pprint(collections.Counter(bigrams).most_common(number))
count_bigrams(inaugural.words())
```

###Describe the class of strings matched by the following regular expressions.


####[a-zA-Z]+

matches one or more letters (from a-z) independent of case eg. absalon or Absalon

####[A-Z][a-z]*
matches one upper case letter and zero or more lower case letters (from a-z) eg. Absalon or A

####p[aeiou]{,2}t
matches a word with a p, 2 or less vowels, a t e.g. pt, pet, pat, pout
####\d+(\.\d+)?
matches a word with one digit, a group of a dot and one or more digits, which is optional e.g. a number both integer or floating point, 1, 1.1
####([^aeiou][aeiou][^aeiou])*
matches a word that's not starting with a vowel, then a vowel, then not starting with a vowel, repeated zero or more times eg. pet, pat, petpatput
####\w+|[^\w\s]+
Matches a word with one or more word characters OR matches a word consisting of one or more "not" a word characters (punctuation for example) and whitespace characters. e.g. "this is some words" or "w. rx. btrasd,"

###Write a utility function that takes a URL as its argument, and returns the contents of the URL, with all HTML markup removed.

```python
import urllib
import bs4
# nltk's clean_html function is deprecated, using beautifulsoup!
def utility_func(url):
    f = urllib.urlopen(url).read()
    cleaned = bs4.BeautifulSoup(f).get_text()
    return cleaned
utility_func("http://www.jp.dk")
```

###Pig Latin is a simple transformation of English text. Each word of the text is converted as follows: move any consonant (or consonant cluster) that appears at the start of the word to the end, then append ay, e.g. string → ingstray, idle → idleay. http://en.wikipedia.org/wiki/Pig_Latin

###A: Write a function to convert a word to Pig Latin.


```python
def pig_latin_word(word):
    index = 0
    for i, char in enumerate(word):
        if char.lower() in "aeiou":
            index = i
            break
    return word[index:]+word[:index]+"ay"

print(pig_latin_word("string"))
print(pig_latin_word("idle"))
print(pig_latin_word("See"))
print(pig_latin_word("You"))
print(pig_latin_word("Later"))
```

###B: Write code that converts text, instead of individual words.

```python
def pig_latin_text(text):
    return " ".join([pig_latin_word(word) for word in text.split()])

pig_latin_text("String Idle See you later")
    
```

### Use the Porter Stemmer to normalize some tokenized text, calling the stemmer on each word. Do the same thing with the Lancaster Stemmer and see if you observe any differences.

First the PorterStemmer:

```python
stemmer_text = [w.lower() for w in inaugural.words() if w.lower() not in stopwords.words("english") and w.lower().isalnum()]
porter = nltk.PorterStemmer()
pprint([porter.stem(w) for w in stemmer_text])
```

```python
lancaster = nltk.LancasterStemmer()
pprint([lancaster.stem(w) for w in stemmer_text])
```

It seems the Lancaster stemmer is generally more aggressive and cutting the words shorter than the Porter stemmer:

citizen - cit, senat - sen, life - lif, event - ev, and so on

####What is the definition of a social bot?  



Software programs posing as humans on social media sites interacting with humans either conspicuously or nonconspicuously. ####What is the Turing test?

The Turing Test is a "test" envisioned by Alan Turing, and asks if a software program is capable of fooling a person into believing it is human. This is relevant in the context of social bots, because they are software programs trying to pose as humans.

####When can a social bot claim to have passed the Turing test?

It can claim to have passed the Turing test when it interacts with real users and give meaningful inquiries or feedback fooling the users into thinking it is a real person.
