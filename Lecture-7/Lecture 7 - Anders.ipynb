## Problem 1 - Connected components

A connected component is defined as a sub-graph of a super-graph (which is the entire graph). In other words, it's a part of a graph, where this part is a graph it self. <br><br>

### Giant component

A giant connected component is (qualitative observation speaking) usually the biggest sub-graph of the super-graph. That is, the sub-graph which contains the most nodes. 
A probability p must be bigger than a threshold before a giant component can occour. The threshold is defined as 1/(n-1), where n=number of nodes. p is defined the probability that each node "spawn" an edge. 

We can also calculate how big p should be if the whole graph should be completely connected:<br>
p_conn = log(n)/n

## Problem 3 - Clustering coefficient

The clustering coefficient, is a messure of a node A's neighbors, also are connected to each other. You can say it's the probability of the neihbors are aklso connected. Th result is a number between 0 and 1. 
Clustering coefficient can be devided in two:
* Local clustering coefficient
* Global clustering coefficient



**Local clustering coefficient:** <br>
This is for only one node. 
We can calculate the clustering coefficient with this formula: 
$$cc(v)=\frac{number\quad of\quad pairs\quad of\quad neighbors\quad connected\quad by\quad edges}{number\quad of\quad pairs\quad of\quad neighbors}$$

**Global clustering coefficient:** <br>
This is just the average over ALL of the local clustering coefficient in the graph. While the local clustering coefficient tells about one 1, the global tells about the whole network, which can give an indication of how many in the network know each other. The global can be very helpful with especially large networks


#### Example

<u>*The following example and calculations is based on the graph below*</u><br>

We want to calculate the clustering coefficient for the red node v. 

First we want to find the number of pairs of neighbors connected by edges. <br>
Here there are 3 pairs: (1-3, 2-4, 4-5)

Then, we want to find the number of pairs of neighbors. When we knows them both, we can use the above formula:
$$cc(v) = \frac{3}{\begin{pmatrix} 5\\2 \end{pmatrix}} = \frac{3}{10} = \underline{\underline{0.3}}, \quad where \begin{pmatrix} 5\\2 \end{pmatrix} = \begin{pmatrix} x\\y \end{pmatrix} = \frac{x!}{(y+1)!*y!}  = \frac{5!}{3!*2!}$$
