### 2. Network Analysis B

Using "Wikipedia who-votes-on-whom network" data and printing the 200 first votes.

```python
with open("rfa_all.NL-SEPARATED.txt","r") as f:
    lines = f.readlines()
stripped = [l.strip() for l in lines if l.startswith("SRC") or l.startswith("TGT")]
tuples = []
for i in range(0,len(stripped),2):
    src = stripped[i].split(":")[1]
    tgt = stripped[i+1].split(":")[1]
    tuples.append((src,tgt))
G = nx.DiGraph()
G.add_edges_from(tuples)
```

```python
G_200 = nx.DiGraph()
G_200.add_edges_from(tuples[:200])
plt.figure(3,figsize=(20,20)) 
nx.draw(G_200, with_labels=True)
plt.show()
```

####Computing the average shortest path length:

```python
print(nx.average_shortest_path_length(G))
```

#### Computing the average clustering coefficients:

```python
print(nx.average_clustering(G))
```

#### Computing the degree distribution (using logarithmic scales for the axes):

```python
in_degrees = G.in_degree()
out_degrees = G.out_degree()
in_values = sorted(set(in_degrees.values()))
out_values = sorted(set(out_degrees.values()))
in_hist = [in_degrees.values().count(x) for x in in_values]
out_hist = [out_degrees.values().count(x) for x in out_values]
plt.figure(3,figsize=(20,20))
plt.plot(in_values,in_hist,'ro-')
plt.yscale("log")
plt.xscale("log")
plt.plot(out_values,out_hist,'bv-') 
plt.legend(['In-degree','Out-degree'])
plt.xlabel('Degree')
plt.ylabel('Number of nodes')
plt.title('Wikipedia who-votes-on-whom network data')
plt.show()
```

### And be careful - some of the networks on snap are large, and the average path length may take a long time to calculate for big graphs. Why is that the case?

Because for every node, it has to compute the path lengths to all other nodes, so N-1 path lengths for every node.

So $(N \cdot (N-1))$ calculations for directed graphs or $(N \cdot (N-1))/2$ for undirected graphs.
