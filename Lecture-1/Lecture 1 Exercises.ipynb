### Why is JSON superior to XML
*JSON pros*
*    Simple notation
*    lightweightness that leads to it being performant
*    maps/deserializes better to data structures (lists, dictionaries)

*XML pros*
*    is a document markup language that enables it to give additional meaning to data


```python
import json
import urllib2
import pprint

url = "http://dl.dropbox.com/u/153071/test.json"
response = urllib2.urlopen(url)
urlfile = response.read()
response.close()
jsonfile = json.loads(urlfile)
pprint.pprint(jsonfile)
```
