```python
#Directed graphs
dg = nx.DiGraph()
dg.add_weighted_edges_from([(1,4,0.5), (3,1,0.75)])
n = 1
print "Total weight FROM the node '%r': %r" % (n, dg.out_degree(n, weight="weight"))
print "Total weight TO the node '%r': %r" % (n, dg.degree(n,weight="weight"))
dg.add_weighted_edges_from([(1,5,0.5), (2,1,0.75)])
print "List of nodes that '%r' pointing to: %r" % (n, dg.successors(n))
print "List of nodes which pointing TO '%r': %r" % (n, dg.predecessors(n))
```

```python
#Multigraphs - MultiGraph and MultiDiGraph.

mg = nx.MultiGraph()
mg.add_weighted_edges_from([(1,2,.5), (1,2,.75), (2,3,.5)])
print mg.degree(weight="weight")
```

### Fetching the followers of followers as nodes and their relationships as edges

```python
import twitter
import time

with open("../bot_config.txt","r") as f:
    config = f.readlines()
config_dict = dict(line.strip().split("=") for line in config if not line.startswith("#"))

CONSUMER_KEY = config_dict["CONSUMER_KEY"]  # API key
# API secret key
CONSUMER_SECRET = config_dict["CONSUMER_SECRET"]
# Access token
OAUTH_TOKEN = config_dict["OAUTH_TOKEN"]
# access token secret
OAUTH_TOKEN_SECRET = config_dict["OAUTH_TOKEN_SECRET"]

auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET, CONSUMER_KEY, CONSUMER_SECRET)
twitter_api = twitter.Twitter(auth=auth)

print twitter_api
```

```python
result = twitter_api.followers.list(screen_name="JackBoHorseMan", count=200)
edges = []
count = 1
nodes = []
nodes.append("JackBoHorseMan")
for user in result["users"][:14]:
    print count
    result2 = twitter_api.followers.list(screen_name=user["screen_name"], count=50)
    edges.append(("JackBoHorseMan", user["screen_name"]))
    nodes.append(user["screen_name"])
    for user2 in result2["users"]:
        edges.append((user["screen_name"], user2["screen_name"]))
        nodes.append(user2["screen_name"])
    count +=1

#print tuplelist
print "Good night..."
time.sleep((60*15)+1)
print "Good morning..."
count = 1
for user in result["users"][15:29]:
    result2 = twitter_api.followers.list(screen_name=user["screen_name"], count=50)
    edges.append(("JackBoHorseMan", user["screen_name"]))
    nodes.append(user["screen_name"])
    for user2 in result2["users"]:
        edges.append((user["screen_name"], user2["screen_name"]))
        nodes.append(user2["screen_name"])
    count +=1

nodes = list(set(nodes))
```

### Saving the nodes and the edges in files

```python
File = open('twitterGraphNodes.txt', 'w')
[File.write(item+"\n") for item in nodes]
File.close()

File = open('twitterGraphEdges.txt', 'w')
[File.write("("+x+","+y+")\n") for x,y in edges]
File.close()
```

### Loading the nodes and edges from files

```python
def load(Filename, tupleFlag=0):
    line = []
    File = open(Filename, 'r')
    lines = File.readlines()
    File.close()
    if tupleFlag == 0:
        List = [item.rstrip() for item in lines]
        return List
    else:
        tupleList = []
        for line in lines:
            itemSet = line.split(",")
            tupleList.append((itemSet[0],itemSet[1].rstrip()))
        return tupleList


nodes = load('twitterGraphNodes.txt')
edges = load('twitterGraphEdges.txt', 1)
print "Done! :)"
```

### Adding the nodes and edges to graph and draw it

```python
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline

tg = nx.Graph()
tg.add_nodes_from(nodes)
tg.add_edges_from(edges)
```

```python
plt.figure(figsize=(15,13));
nx.draw(tg)
```

### Writing graph to a graphml-file

```python
nx.write_graphml(tg, "twitterGraph.graphml")
```
