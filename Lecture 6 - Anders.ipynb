```python
#Accessing nodes and edges - part 1
g.add_node('a')
print "Number of nodes: %d" % g.number_of_nodes() #or g.order()
print "Number of edges: %d" % g.number_of_edges() #or g.size()

print "The nodes: %r" % g.nodes() #Printing all the nodes in the graph g
print "The edges: %r" % g.edges() #Printing all the edges in the graph g

n=1
print "The neighbor nodes to '%r': %r" % (n, g.neighbors(n)) #Printing the neighbors to the node '1'
print "The number of neigbor nodes to '%r': %r" % (n, g.degree(n))
```

```python
#Directed graphs
dg = nx.DiGraph()
dg.add_weighted_edges_from([(1,4,0.5), (3,1,0.75)])
n = 1
print "Total weight FROM the node '%r': %r" % (n, dg.out_degree(n, weight="weight"))
print "Total weight TO the node '%r': %r" % (n, dg.degree(n,weight="weight"))
dg.add_weighted_edges_from([(1,5,0.5), (2,1,0.75)])
print "List of nodes that '%r' pointing to: %r" % (n, dg.successors(n))
print "List of nodes which pointing TO '%r': %r" % (n, dg.predecessors(n))
```

```python
#Multigraphs - MultiGraph and MultiDiGraph.

mg = nx.MultiGraph()
mg.add_weighted_edges_from([(1,2,.5), (1,2,.75), (2,3,.5)])
print mg.degree(weight="weight")
```

```python
#Draw a graph
import pylab as plt #import Matplotlib plotting interface
g = nx.erdos_renyi_graph(100,0.15)
nx.draw(g)
nx.draw_random(g)
nx.draw_circular(g)
nx.draw_spectral(g)
plt.savefig('graph.png')
```

```python
import twitter
import time

with open("bot_config.txt","r") as f:
    config = f.readlines()
config_dict = dict(line.strip().split("=") for line in config if not line.startswith("#"))

CONSUMER_KEY = config_dict["CONSUMER_KEY"]  # API key
# API secret key
CONSUMER_SECRET = config_dict["CONSUMER_SECRET"]
# Access token
OAUTH_TOKEN = config_dict["OAUTH_TOKEN"]
# access token secret
OAUTH_TOKEN_SECRET = config_dict["OAUTH_TOKEN_SECRET"]

auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET, CONSUMER_KEY, CONSUMER_SECRET)
twitter_api = twitter.Twitter(auth=auth)

print twitter_api
```

```python
result = twitter_api.followers.list(screen_name="JackBoHorseMan", count=200)
tuplelist = []

for user in result["users"][:10]:
    result2 = twitter_api.followers.list(screen_name=user["screen_name"], count=50)
    tuplelist += ("JackBoHorseMan", user["screen_name"])
    for user2 in result2["users"]:
        tuplelist += (user["screen_name"], user2["screen_name"])

#print tuplelist
print "Good night..."
time.sleep((60*15)+1)
print "Good morning..."
for user in result["users"][10:]:
    result2 = twitter_api.followers.list(screen_name=user["screen_name"], count=50)
    tuplelist += ("JackBoHorseMan", user["screen_name"])
    for user2 in result2["users"]:
        tuplelist += (user["screen_name"], user2["screen_name"])

```

```python
import pickle

File = open("twitterGraph.txt", 'w')

pickle.dump(tuplelist, File)

File.close()
```
